Task 5: Data Analysis

**Univariate Analysis**

```python
#Price distribution
data["price"]
```

```python
# calculate quartiles
Q1 = data["price"].quantile(0.25)
Q3 = data["price"].quantile(0.75)

IQR = Q3 - Q1

# bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print("Q1:", Q1)
print("Q3:", Q3)
print("IQR:", IQR)
print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)
```

The interquartile range (IQR) for price is 119, with Q1 at 80 and Q3 at 199. The upper bound for outliers is 377.5, meaning listings priced above this value are considered outliers. Since the lower bound is negative and prices cannot be negative, no lower outliers exist. Therefore, only high-priced listings above 377.5 are treated as outliers.

```python
plt.figure(figsize=(8,5))
sns.boxplot(x=data["price"])

plt.xlim(0, 1000)   
plt.show()
```

```python
outliers = data[(data["price"] < lower_bound) | (data["price"] > upper_bound)]

print("Number of outliers:", outliers.shape[0])
```

```python
df_clean = data[(data["price"] >= lower_bound) & (data["price"] <= upper_bound)]
```

```python
plt.figure(figsize=(8,5))

sns.histplot(data=df_clean, x="price", bins=50)

plt.title("Price Distribution (Without Outliers)")
plt.xlabel("Price")
plt.ylabel("Frequency")

plt.show()
```

The price distribution is highly right-skewed, with most listings concentrated between 80 and 200. A few high-priced listings act as outliers, indicating the presence of premium properties

```python
data.columns
```

```python
plt.figure(figsize=(6,3))

sns.histplot(data=data, x="availability_365", bins=30)

plt.title("Availability (365 days) Distribution")
plt.xlabel("Availability (days)")
plt.ylabel("Frequency")

plt.show()
```

Availability is concentrated at extreme values (0 and 365), suggesting that listings are either fully booked or completely available. This indicates uneven demand across properties

**Bi variate Analysis**

```python
data = data[data["beds"] > 0]
```

```python
data["price_per_bed"] = data["price"] / data["beds"]
```

```python
data[["price", "beds", "price_per_bed"]].head()
```

```python
df_clean = data[(data["price"] >= lower_bound) & (data["price"] <= upper_bound)]
df_clean.groupby("neighbourhood_group")["price_per_bed"].mean()
```

Average price per bed for each neighbourhood group

```python
print(df_clean.groupby("neighbourhood_group")["price"].mean())
```

Average price of listings in each neighbourhood group

```python
data.columns
```

```python
# price dependency on neighbourhood
sns.barplot(data=df_clean, x="neighbourhood_group", y="price", hue="room_type")
```

Prices vary significantly across neighbourhoods, with Manhattan being the most expensive. Entire homes consistently command higher prices compared to private and shared rooms.

```python
# number of reviews and price relationship
plt.figure(figsize=(8,5))
plt.title("Locality and Review Dependency")
sns.scatterplot(data=df_clean, x="number_of_reviews", y="price", hue="neighbourhood_group")
plt.show()
```

An inverse relationship is observed between price and number of reviews, indicating that lower-priced listings tend to receive more reviews and higher engagement.

```python
data.dtypes
```

```python
sns.pairplot(
    data=df_clean,
    vars=["price", "minimum_nights", "number_of_reviews", "availability_365"],
    hue="room_type")
```

The pairplot reveals that price is right-skewed and inversely related to the number of reviews, indicating that more affordable listings tend to attract higher demand. Availability also shows a negative relationship with reviews, suggesting that frequently booked listings remain less available. Most other variables exhibit weak relationships, highlighting the complexity of factors influencing Airbnb pricing and demand.

```python
#Geographical Distribution of AirBnb Listing
plt.figure(figsize=(10,7))

sns.scatterplot(
    data=data,
    x="longitude",
    y="latitude",
    hue="room_type",
    alpha=0.5)

plt.title("Geographical Distribution of Airbnb Listings")

plt.show()
```

Listings are densely concentrated in central locations, indicating high demand areas. Room type distribution varies across regions, with entire homes dominating premium zones.

```python
#heat map - correlation of one variable with others for numerical column

corr = df_clean[[
    "latitude", "longitude", "price", "minimum_nights",
    "number_of_reviews", "reviews_per_month",
    "availability_365", "beds"]].corr()

plt.figure(figsize=(8,6))
sns.heatmap(data=corr, annot=True)

plt.title("Correlation Heatmap")
plt.show()
```

Price shows a moderate positive correlation with number of beds, indicating larger properties are more expensive. Availability is negatively correlated with reviews, suggesting higher demand reduces availability.

**Final Conclusion**

The analysis reveals that price and demand in Airbnb listings are influenced by multiple factors such as location, room type, and property size. Central locations and entire homes command higher prices, while affordable listings tend to receive more bookings. Availability patterns indicate uneven demand distribution, with some listings being consistently booked while others remain underutilized.
